```python
import pandas as pd

df = pd.read_csv("/content/data/Amazon-Products.csv")
df.head()
```

```python
df.columns = df.columns.str.strip().str.lower()
print(df.columns)
```

```python
df = df.drop(columns=["unnamed: 0"])
```

```python
print("Missing Values:\n")
print(df.isnull().sum())
```

```python
df["name"] = df["name"].fillna("Unknown Product")
df["main_category"] = df["main_category"].fillna("Unknown")
df["sub_category"] = df["sub_category"].fillna("Unknown")
```

```python
# remove ₹ and commas
df["discount_price"] = df["discount_price"].replace('[₹,]', '', regex=True)
df["actual_price"] = df["actual_price"].replace('[₹,]', '', regex=True)

# convert to numeric
df["discount_price"] = pd.to_numeric(df["discount_price"], errors='coerce')
df["actual_price"] = pd.to_numeric(df["actual_price"], errors='coerce')

# fill missing values safely
df["discount_price"] = df["discount_price"].fillna(df["discount_price"].median())
df["actual_price"] = df["actual_price"].fillna(df["actual_price"].median())
```

```python
df["ratings"] = df["ratings"].fillna(df["ratings"].mean())
df["no_of_ratings"] = df["no_of_ratings"].fillna(0)
```

```python
df = df.drop_duplicates()
```

```python
df["discount_percent"] = ((df["actual_price"] - df["discount_price"])
                          / df["actual_price"]) * 100
```

```python
print(df["main_category"].value_counts().head())
```

```python
print(df.sort_values(by="actual_price", ascending=False)[["name","actual_price"]].head())
```

```python
print(df.sort_values(by="ratings", ascending=False)[["name","ratings"]].head())
```

```python
print("\nCleaned Dataset Info:\n")
print(df.info())

print("\nRemaining Missing Values:\n")
print(df.isnull().sum())
```
