```python
import math

def calculate_area(radius):
    """Returns area of a circle"""
    area = math.pi * radius * radius
    return area

r = float(input("Enter radius: "))
print("Area of circle =", round(calculate_area(r), 2))
```

```python
def find_largest(a, b, c):
    largest = a

    if b > largest:
        largest = b
    if c > largest:
        largest = c

    return largest

print("Largest number is:", find_largest(12, 45, 21))
```

```python
def string_length(text):
    count = 0
    for char in text:
        count += 1
    return count

user_text = input("Enter a string: ")
print("Length:", string_length(user_text))
```

```python
def check_prime(num):
    if num <= 1:
        return "Not Prime"

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return "Not Prime"

    return "Prime"

n = int(input("Enter number: "))
print(check_prime(n))
```

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

num = int(input("Enter position: "))
print("Fibonacci number is:", fibonacci(num))
```

```python
def reverse_text(text):
    if text == "":
        return text
    return reverse_text(text[1:]) + text[0]

word = input("Enter a word: ")
print("Reversed:", reverse_text(word))
```

```python
n = int(input("Enter a number: "))

total = 0
for i in range(1, n + 1):
    total += i

print("Sum from 1 to", n, "=", total)
```

```python
numbers = [3, 8, 11, 14, 20, 25, 30]

even_total = 0

for value in numbers:
    if value % 2 == 0:
        even_total += 1

print("Total even numbers:", even_total)
```

```python
def is_prime(num):
    if num < 2:
        return False

    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

print("Prime numbers from 1 to 100:")

for number in range(1, 101):
    if is_prime(number):
        print(number, end=" ")
```
